```python
# Ввод необходимых библиотек
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import enum 
import scipy.optimize as optimize
import datetime
import statsmodels.api as sm
import pandas as pd
import time
import pickle
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
```

```python
#names = ['BS', 'Heston', 'Bates', 'Bates_mod']
names = ['BS', 'Heston', 'Bates', 'Bates_mod']
#assets = ['CVX', 'UNH', 'BA', 'NVDA', 'QCOM', 'AMD', 'NVDAQCOM', 'BACVX', 'NVDAAMDQCOM', 'UNHCVXBA']
assets = ['NVDAQCOM', 'BACVX', 'NVDAAMDQCOM', 'UNHCVXBA']
```

```python
def gra(assets, names):
  for asset in assets:
    fig1, ax1 = plt.subplots(figsize=(5, 4))
    fig2, ax2 = plt.subplots(figsize=(5, 4))
    for name in names:
      x = []
      y1 = []
      y2 = []
      b_file = open(f"results_{name}.pkl", "rb")
      data = pickle.load(b_file)
      b_file.close()
      #print(data[asset])
      for i in data[asset]:
        x.append(i[0])
        y1.append(i[1])
        y2.append(i[2])
      x = np.array(x)
      y1 = np.array(y1)
      y2 = np.array(y2)
      ax1.plot(x, y1, label=name)
      ax1.plot(x, y1, 'o')
      ax1.set_xlim(-0.5, 10.5)
      ax2.plot(x, y2, label=name)
      ax2.plot(x, y2, 'o')
      ax2.set_xlim(-0.5, 10.5)
    #ax1.title(asset)
    #ax1.ylabel('Value')
    ax1.set_title(asset + ' Value')
    ax1.legend(loc='upper left', ncol=2)
    #ax2.title(asset)
    #ax2.ylabel('p')
    ax2.set_title(asset + ' p')
    ax2.legend(loc='lower left', ncol=2)
    plt.show()
```

```python
#names = ['BS', 'Heston', 'Bates', 'Bates_mod']
names1 = ['BS', 'Heston', 'Bates']
names2 = ['BS', 'Heston', 'Bates', 'Bates_mod']
assets1 = ['CVX', 'UNH', 'BA', 'NVDA', 'QCOM', 'AMD']
assets2 = ['NVDAQCOM', 'BACVX', 'NVDAAMDQCOM', 'UNHCVXBA']
assets3 = ['NVDAAMD', 'QCOMAMD', 'BAUNH' , 'CVXUNH']
#gra(assets1, names1)
gra(assets3, names2)
```

```python
x = np.arange(10)
y = np.sin(x)
cs = CubicSpline(x, y)
xs = np.arange(-0.5, 9.6, 0.1)
fig, ax = plt.subplots(figsize=(6.5, 4))
ax.plot(x, y,  label='data')
ax.plot(xs, np.sin(xs), label='true')
ax.plot(xs, cs(xs), label="S")
ax.plot(xs, cs(xs, 1), label="S'")
ax.plot(xs, cs(xs, 2), label="S''")
ax.plot(xs, cs(xs, 3), label="S'''")
ax.set_xlim(-0.5, 10.5)
ax.legend(loc='lower left', ncol=2)
plt.show()
```
